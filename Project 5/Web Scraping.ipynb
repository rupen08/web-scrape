```python
import requests
from bs4 import BeautifulSoup
import html5lib
import time
import datetime
```

```python
url = "https://codewithharry.com"
```

```python
r = requests.get(url)
```

```python
htmlContent = r.content
```

### Loading the whole page 

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify())
```

```python
## find all the paragraphs

paras = soup.find_all("p")

print(list(paras))
```

```python
## Same, we can print all the headers

headers = soup.find_all("h")

print(list(headers))

## says there is no h tags!!
```

```python
print(soup.find("p")) ## this gives first mentioned element
```

```python
print(soup.find("p")["class"]) ## gives all the classes for mentioned element
```

```python
print(soup.find_all("p", class_ = "lead"))
```

```python
print(soup.find_all("p", class_ = "text-muted"))
```

```python
print(soup.find("p").get_text())
```

```python
## .get_text() method return all the inserted text from the soup file or mentioned element

print(soup.get_text())
```

### Scraping Navigation Bar


```python
divs = soup.find("div")["class"]

list(divs)
```

```python
ids = soup.find("div")["id"]
ids
```

```python
navbar = soup.find(id = "navbarSupportedContent")
navbar
```

```python
for child in navbar.children:
    print(child)
    
## all elements stored as generators, does not occupies any memory
```

```python
for content in navbar.contents:
    print(content)
    
## all elements stored in actual memory
```

```python
for string in navbar.strings:
    print(string)
    
## strings are generallt refered as written between <a> HERE </a> in <div>
```

```python
for string in navbar.stripped_strings:
    print(string)

## strings are generallt refered as written between <a> HERE </a> in <div>
```

```python
aa= soup.find("div", id = "navbarSupportedContent")

aa
```

```python
print(aa.strings)
```

```python
for i in aa.strings:
    print(i)
```

### Scraping the Upper portion of the Website

```python
## This is also called CSS Selector method!

intro = soup.find("div", {"class":"container"})

print(intro.prettify())
```

#### Tip: If want to iterate over multiple tags, use FOR LOOP with FIND_ALL function and THEN USE .GET method to pull out ATTRIBUTE  !

```python
## Printing out just strings

l = []
for string in intro.strings:
    l.append(string)
print(l)
```

```python
intro.get_text()

## gives all text in unstructured data form
## prefer strings for loop instead
```

```python
## Scrap out only word which are readale to viewers on website

for i in intro.find_all("a"):
    print(i.string)
```

```python
## Scrap out all the important Links with the notation!!

for i in intro.find_all("a"):
    print(i.string+": "+url+i.get("href"))
```

```python
print(intro.find(["h1","h2"]).string)

## checking h1, h2 content!!
```

```python
for i in intro.find_all("p"):
    print(i.get_text())
    
## Scrap out text from all paragraphs
```
